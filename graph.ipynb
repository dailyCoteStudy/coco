```python
import matplotlib.pyplot as plt

# 주어진 x, y 좌표
coordinates = [(219, 220), (195, 204), (179, 209), (158, 215), (141, 216), (125, 213), 
               (109, 211), (94, 210), (79, 206), (58, 198)]

# x와 y 좌표 분리
x_values = [coord[0] for coord in coordinates]
y_values = [coord[1] for coord in coordinates]

# 그래프 그리기 (연결선 없이 마커만 표시)
plt.plot(x_values, y_values, marker='o', linestyle='', color='b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X-Y Coordinate Plot')
plt.grid(True)

# y축 범위 설정
plt.ylim(190, 225)

# 각 점 아래에 122, 123, 124, ... 번호 추가
for i, (x, y) in enumerate(coordinates):
    plt.text(x, y - 2, str(122 + i), ha='center', fontsize=9)  # 각 좌표 아래에 번호 추가

plt.show()
```

```python
import matplotlib.pyplot as plt

# 주어진 x, y 좌표
coordinates = [(219, 220), (195, 204), (179, 209), (158, 215), (141, 216), (125, 213), 
               (109, 211), (94, 210), (79, 206), (58, 198)]

# x와 y 좌표 분리
x_values = [coord[0] for coord in coordinates]
y_values = [coord[1] for coord in coordinates]

# 그래프 그리기 (연결선 없이 마커만 표시, 마커 크기를 키움)
plt.plot(x_values, y_values, marker='o', linestyle='', markersize=8, color='b')
plt.xlabel('x-coordinate')
plt.ylabel('y-coordinate')
plt.title('Center point shift')
plt.grid(True)

# y축 범위 설정
plt.ylim(190, 225)

# 각 점 아래에 122, 123, 124, ... 번호 추가 (굵은 글씨로 표시)
for i, (x, y) in enumerate(coordinates):
    plt.text(x, y - 2, str(122 + i), ha='center', fontsize=8, fontweight='bold')  # 글씨를 굵게 추가

# 좌표들 사이에 화살표 그리기
for i in range(len(coordinates) - 1):
    plt.annotate('', xy=coordinates[i + 1], xytext=coordinates[i],
                 arrowprops=dict(facecolor='black', shrink=0.3, width=0.5, headwidth=4, headlength=4))

plt.show()
```

```python
import matplotlib.pyplot as plt

# 주어진 x, y 좌표
coordinates = [(283, 103), (286, 106), (282, 105), (285, 105), (280, 107), (285, 106), 
               (283, 105), (279, 104), (286, 102), (280, 103)]

# 좌표를 x, y 리스트로 분리
x_values = [coord[0] for coord in coordinates]
y_values = [coord[1] for coord in coordinates]

# 그래프에 좌표 찍기
plt.figure(figsize=(10,6))  # 가로 10, 세로 6으로 설정
plt.scatter(x_values, y_values, color='blue')

# 각 좌표에 순번 추가
for i, (x, y) in enumerate(coordinates):
    plt.text(x, y, f'{i+1}', fontsize=12, ha='right', va='bottom', color='red')

# 첫 좌표를 기준으로 반지름이 10인 원 그리기
first_coordinate = coordinates[0]
circle = plt.Circle(first_coordinate, 9, color='gray', fill=False, linestyle='--', linewidth=2)
plt.gca().add_patch(circle)

# x축, y축 설정
plt.xlim(250, 315)  # x축 범위 설정
plt.ylim(92, 114)   # y축 범위 설정

# y축을 5단위로 설정
plt.yticks(range(90, 115, 5))  # y축을 5단위로 설정

plt.title("Center point shift")
plt.xlabel("x-coordinate")
plt.ylabel("y-coordinate")
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # 원의 비율을 유지
plt.show()
```
